```python
from kp_hmm import HMM
from kp_data_loader_no_t import KPDataLoader, sequence
from sklearn.model_selection import KFold, LeaveOneOut
import logging
import sys
from seqlearn import hmm, perceptron
import numpy as np

logger = logging.getLogger()
logging.basicConfig(filename='../../../results/run_one_hmm.log',level=logging.DEBUG)

# -----------------------------------------------

loader = KPDataLoader()
for i in range(1,46):
    loader.load_file('ex{0}a.mid.csv'.format(i))
XX, Y = loader.get_XX_and_YY()
n = len(XX)
#
# # -----------------------------------------------
#
# n = len(XX)
# j = int(n - (float(n) * 0.1))
#
# XX_train = XX[0:j]
# Y_train = Y[0:j]
#
# XX_test = XX[j:n]
# Y_test = Y[j:n]
#
# # -----------------------------------------------

# model = HMM()
# model.train(XX_train, Y_train)
# accuracy, y_pred = model.test(XX_test, Y_test)

# logger.info("Accuracy: {0}".format(accuracy))

def cross_val(XX, Y, n=10):
    """
    n : n-crossvalidation
    """

    L = len(XX)
    kf = KFold(n_splits=n)

    models = []
    scores = []

    c = 0

    for c, (train_indexes, val_indexes) in enumerate(kf.split(XX)):

        logger.debug("On Fold " + str(c))

        xx_train = []
        y_train = []
        xx_val = []
        y_val = []
        
        for i in train_indexes:
            xx_train.append(XX[i][:])
            y_train.append(Y[i][:])
        X_train, Y_train, L_train = sequence(xx_train, y_train)
        for j in val_indexes:
            xx_val.append(XX[j][:])
            y_val.append(Y[j][:])
        X_val, Y_val, L_val = sequence(xx_val, y_val)

        model = hmm.MultinomialHMM()

        logger.debug(str(len(X_train)) + "," + str(len(Y_train)))
        model.fit(X_train, Y_train, L_train)

        logger.debug("Testing ...")
        accuracy = model.score(X_val, Y_val, L_val)

        logger.debug("Fold " + str(c) + " scored " + str(accuracy))

        models.append(model)
        scores.append(accuracy)

#     max_score = max(scores)

#     print max_score

#     max_index = 0
#     for idx, score in enumerate(scores):
#         if score == max_score:
#             max_index = idx
#             break
    
    return np.mean(scores)

# 		logger.info("Final Test ...")

# 		score, _ = models[max_index].test(XX_test,Y_test)
# 		logger.info("Final Accuracy: {0}".format(score))

# 		return models[max_index]

mean = cross_val(XX, Y, n)
print "Mean:", mean

# -----------------------------------------------

# Now with seqlearn

# logger.info("Seqlearn time ... ")
# # Get data again just in case I messed up somewhere
# loader = KPDataLoader()
# for i in range(1,46):
#     loader.load_file('ex{0}a.mid.csv'.format(i))
# XX, YY = loader.get_XX_and_YY()

# # -----------------------------------------------

# n = len(XX)
# j = int(n - (float(n) * 0.5))

# XX_train = XX[0:j]
# YY_train = YY[0:j]

# X_train, Y_train, L_train = sequence(XX_train, YY_train)

# XX_test = XX[j:n]
# YY_test = YY[j:n]

# X_test, Y_test, L_test = sequence(XX_test, YY_test)

# logger.info('Building HMM model ...')
# model = hmm.MultinomialHMM()
# logger.info('Training model ...')
# model.fit(X_train, Y_train, L_train)
# logger.info('Testing model ...')
# accuracy = model.score(X_test, Y_test, L_test)

# logger.info("Accuracy: {0}".format(accuracy))
# for i in model.predict(X_test, L_test):
#     print "{0} ".format(i),

# logger.info('Building Percpetron model ...')
# model = perceptron.StructuredPerceptron()
# logger.info('Training model ...')
# model.fit(X_train, Y_train, L_train)
# logger.info('Testing model ...')
# accuracy = model.score(X_test, Y_test, L_test)
# logger.info("Accuracy: {0}".format(accuracy))
# -----------------------------------------------
```
